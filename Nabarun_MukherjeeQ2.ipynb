```python
import numpy as np
import pandas as pd
```

**Answer to Question No 2 (a)**

```python
def sigmoid(z):
  return 1/(1+ np.exp(-z))

class LogisticRegression:
  def __init__(self, learning_rate=0.01, num_iterations=1000):
    self.learning_rate=learning_rate
    self.num_iterations=num_iterations

  def fit(self, X, y):
    m,n=X.shape # n is the number of coulmns ie the no of parameters and m is the number of samples
    self.theta=np.zeros(n) #theta array has all the parameters and all are set to zero and then gradient descent will be used to find the argmin(parameters)

    for _ in range(self.num_iterations):
      z= np.dot(X,self.theta)
      h=sigmoid(z)
      gradient = np.dot(X.T, (h-y))/m
      self.theta= self.theta - (self.learning_rate * gradient)

  def predict(self , X):
    z=np.dot(X, self.theta)
    h=sigmoid(z)
    return (h>=0.5).astype(int)
```

```python
file= "football.csv"
df=pd.read_csv(file)
df
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=categorical_columns, prefix=categorical_columns, prefix_sep='_')
df.fillna(0, inplace=True)
df.columns
```

```python
#splitting the data into training and testing set
import random
split_ratio=0.8
m,n=df.shape
num_samples=m
num_training_samples=int(split_ratio*num_samples)
num_testing_samples= num_samples - num_training_samples

random.shuffle(df.values) #shuffle the data only by values

training_data=df[:num_training_samples]
testing_data=df[num_training_samples:]
X_train= training_data.drop(['contribution_type'], axis=1)
y_train= training_data['contribution_type']
X_test= testing_data.drop(['contribution_type'], axis=1)
y_test= testing_data['contribution_type']
```

**Answer to Question No 2 (b)**

```python
model = LogisticRegression(learning_rate=0.01, num_iterations=1000)
model.fit(X_train,y_train)
y_predict= model.predict(X_test)
```

**Answer to Question No 2 (c)**

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
accuracy = accuracy_score(y_test, y_predict)
f1 = f1_score(y_test, y_predict)

print("Accuracy:", accuracy)
print("F1 Score:", f1)

conf_matrix = confusion_matrix(y_test, y_predict)
print("Confusion Matrix:")
print(conf_matrix)
```
